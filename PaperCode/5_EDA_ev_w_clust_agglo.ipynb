# 5. Process Analytics - EDA Eventlog

```r
# install packages
# install.packages("readxl")
# install.packages("bupaR")
# install.packages("edeaR")
# install.packages("processmonitR")
# install.packages("processmapR")
```

```r
# load libraries
library(processmapR)
library(readxl)
library(bupaR)
```

```r
# load event data
event_data <- read.csv("ev_prep_Activity_Aglo_Avg.csv")
```

```r
str(event_data)
```

```r
# convert to suitable time format
event_data <- convert_timestamps(event_data, columns="time.timestamp", format = dmy_hms)
```

```r
log <- event_data %>%
        eventlog(
            case_id = "case_id",
            activity_id = "Activity_Aglo_Avg",
            activity_instance_id = "activity_instance_new",
            lifecycle_id = "status",
            timestamp = "time.timestamp",
            resource_id = "resource"
        )
```

```r
dim(log)
```

```r
mapping(log)
```

## EDA Eventlog

### Activities

```r
# How many distint activities are in the event log?
n_activities(log)
```

```r
# What are the unique activities in the event log?
activity_labels(log)
```

```r
# Frequency of activities 
activities(log)
```

```r
# calculate summary of activities per process instance (min, max, avg)
table(log$case_id)
```

```r
log_df <- data.frame(table(log$case_id))
```

```r
min(log_df[,2])
```

```r
max(log_df[,2])
```

```r
mean(log_df[,2])
```

```r
# exploring the activitiy presence: 
# activity presence shows in what percentage of process instances an activity is present
log %>% activity_presence() %>%
    plot
```

```r
# Which cases had activities duplicated? 
# The duplicated activities were interrupted by other activities and these duplicated activities were redone by a different resource.
# number_of_selfloops(log, level="case", type = "redo")
```

### Process Instances

```r
# How many orders are recorded in the event log?
n_cases(log)
```

```r
traces(log)
```

```r
# process variants
nrow(traces(log))
```

### Process Mining

```r
process_map(log, type = frequency(value = "absolute_case"))
```

```r
# out <- capture.output(log %>% process_map(type = frequency(value = "absolute_case")))
```

```r
get_flows(process_map(log))
```

```r
# Get data values for activities and flows from process map
get_activities(process_map(log))
```

```r
# create trace coverage graph
log %>%
  trace_coverage(level="trace") %>%
  plot()
```

```r
trace_explorer(log, 0.9)
```
